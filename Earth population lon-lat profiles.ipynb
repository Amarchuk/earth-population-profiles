```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import pandas as pd
import seaborn as sns
# from matplotlib.ticker import MultipleLocator

%matplotlib inline
```

```python
data = pd.read_csv('SEDAC_POP_2000-01-01_rgb_720x360.csv')
```

```python
data.describe()
```

```python
fig = plt.figure(figsize=(10, 6))
ax = sns.heatmap(np.log(data))
ax.set_yticks([])
ax.set_xticks([])
plt.show()
```

Replace water special value with zero:

```python
data = data.replace(99999.0, 0.)
```

```python
fig = plt.figure(figsize=(10, 6))
plt.imshow((np.log(data)).as_matrix(), cmap=plt.cm.Blues)
plt.colorbar()
plt.show()
```

```python
data.describe()
```

Summ all values by longitude:

```python
lon_sum = data.sum(axis=1)
```

```python
lon_sum.describe()
```

```python
#Bad dataset without absolute values:
lon_sum.sum()
```

```python
lon_sum.plot()
plt.show()
```

```python
lat_sum = data.sum()
```

```python
lat_sum.plot()
plt.show()
```

```python
len(lat_sum)
```

```python
fig = plt.figure(figsize=(15, 12))
fig.subplots_adjust(hspace=0, wspace=0)

ax1 = plt.subplot2grid((5,5), (0,0), colspan=4)
ax1.plot(lat_sum)
plt.setp(ax1.get_xticklabels(), visible=False)

ax2 = plt.subplot2grid((5,5), (1,0), colspan=4, rowspan=4, sharex=ax1)
ax2.imshow((np.log(data)).as_matrix(), cmap='Reds', interpolation='nearest', aspect='auto')


ax3 = plt.subplot2grid((5,5), (1, 4), rowspan=4, sharey=ax2)
ax3.plot(lon_sum.values, lon_sum.index)
ax3.set_ylim(len(lon_sum), 0)
plt.setp(ax3.get_yticklabels(), visible=False)

plt.show()
```
